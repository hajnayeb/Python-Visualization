```python
# Import required libraries
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Reading the airline data from a CSV file
airline_data =  pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv', 
                            encoding = "ISO-8859-1",dtype={'Div1Airport': str, 'Div1TailNum': str,'Div2Airport': str, 'Div2TailNum': str})
airline_data.head(3)
```

```python
# Data dimensions:
airline_data.shape
```

```python
# Selecting random but reproducable 500 samples from the data:
d500 = airline_data.sample(n=500, random_state=2)
```

```python
d500.head(3)
```

```python
# Checking the dimensions:
d500.shape
```

```python
# Creating a figure using "go.Figure" and adding trace to it through go.scatter
fig = go.Figure(data=go.Scatter(x=d500['Distance'], y=d500['DepTime'], mode='markers', marker=dict(size=4, color='blue')))
# We can update the figure by adding title and showing labels on x and y axes:
fig.update_layout(title='Distance vs. Departure Time', xaxis_title='Distance', yaxis_title='Departure Time')
# Showing the figure:
fig.show()
```

```python
# Grouping the data based on Month and then computing the averaged arrival delay time:
md = d500.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
# Display the data
md.head(3)
```

```python
# Creating a new figure to visualize the new data
fig = go.Figure(data=go.Scatter(x=md['Month'], y=md['ArrDelay'], mode='markers+lines', marker=dict(size=6, color='red')))
# labels on x and y axes:
fig.update_layout(xaxis_title='Month no.', yaxis_title='Arrival delay time')
# Showing the figure:
fig.show()
```

```python
# Grouping the data by the destination state and reporting the total number of flights for those
bd = d500.groupby(['DestState'])['Flights'].sum().reset_index()
```

```python
# A preview of the data
bd.head(3)
```

```python
# Plotting the data in a bar chart:
fig = px.bar(bd, x="DestState", y="Flights", title = 'total flights of each state') 
fig.show()
```

```python
# Plotting a bubble chart for the previous data
fig = px.scatter(bd, x="DestState", y="Flights", size="Flights", title = 'total flights of each state', size_max=60)
fig.show()
```
